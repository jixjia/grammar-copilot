```python
import os
import requests
import json
import textwrap
import openai
from utilities import config

def pprint(text):
    return print(textwrap.fill(text, width=120))

def get_aoai_response(response):
    output = response['choices'][0]['message']['content'].lstrip()
    print('Finish: ', response['choices'][0]['finish_reason'])
    print('Model: ', response['model'])
    print('Type: ', response['object'])
    print('Prompt Tokens: ', response['usage']['prompt_tokens'])
    print('Completion Tokens: ', response['usage']['completion_tokens'])
    print('Output:\n', output)
    return output
```

```python
# GPT-4 / ChatCompletion
openai.api_type = 'azure'
openai.api_version = config.GPT4_API_VERSION
openai.api_key = config.GPT4_KEY
openai.api_base =  config.GPT4_ENDPOINT
model_name = config.GPT4_DEPLOYMENT

# openai.api_type = 'azure'
# openai.api_version = '2023-06-01-preview'
# openai.api_key = config.GPT35_API_KEY
# openai.api_base =  config.GPT35_ENDPOINT
# model_name = config.GPT35_DEPLOYMENT


### Mimicing User Inputs

key_topic = 'a thank you letter to the owner of the property in which I will be sleeping over for 1 night'
context = 'The owner rented one of his properties on AirBnB, and I applied to stay there by paying them upfront. '
draft_writing = '''
Hi Linda,

My name is Jixin. I am taking my 3yo boy to the Richmond Vale Railway Museum on the 16-Jul. We would love to spend a night prior over at one of the most beautiful houses I have seen. I'd like to thank you in advance to your kind hosting. Please let us know if there's any special rules we need to be aware of. We will keep it tidy and clean and return it exactly as how it was provided to us. We're very looking forward to this trip and thank you once again.

Yours,
Jixin
'''
tone_style = 'professional'
advanced_vocabulary = 'prefer'
memories = None


systemInsructions = [{
    'role':'system', 
    'content': '''You are an English writing expert. 
                  A user is writing about {key_topic} and below is more context about this writing: \n
                  {context}.\n\n
                  Here is the user's draft: \n
                  {draft_writing}.'''.format(key_topic=key_topic, context=context, draft_writing=draft_writing)
}]

messages = [
    {'role':'user', 'content': f"hi, please review my draft writing above and help me rewrite it to make it less verbose, more concise and reads {tone_style}. I {advanced_vocabulary} to use advanced vocabularies where applicable. Thank you."}
]

chatHistory = []
if memories:    
    for idx, memory in enumerate(memories): 
        chatHistory.append({
            'role': 'assistant' if idx%2==0 else 'user',
            'content': memory
        }) 

response = openai.ChatCompletion.create(
    engine=model_name, 
    messages=systemInsructions+messages+chatHistory, 
    temperature=0.5,
    max_tokens=4024,
    stop=None,
    frequency_penalty=0.0,
    presence_penalty=0.0
    )

get_aoai_response(response)
```
